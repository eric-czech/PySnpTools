```python
import os
from pysnptools.distreader import DistGen, Bgen
import  pandas as pd
from datetime import datetime

def timing(iid_count_list,sid_count_list,bits_list,dtype_list,block_size=None):
    #df = pd.DataFrame(columns=['iid_count', 'sid_count', 'gen','cache','read'])
    
    list_row = []
    file_pattern = r'm:\del35\temp{0}.bgen' #r'/mnt/m/del35/temp{0}.bgen'
    for iid_count in iid_count_list:
        for sid_count in sid_count_list:
            for bits in bits_list:
                val_count = iid_count * sid_count
                key = '{0}x{1}-{2}'.format(iid_count,sid_count,bits)
                file = file_pattern.format(key)
                if not os.path.exists(file):
                    distgen = DistGen(seed=0,iid_count=iid_count,sid_count=sid_count,block_size=block_size)
                    print('{0} gen and write to bgen'.format(key))
                    bgen = Bgen.write(file,distgen,bits=bits)
                    print('{0} cache'.format(key))
                    bgen.iid
                    bgen.flush()
                for dtype in dtype_list:
                    time_c = datetime.now()
                    bgen = Bgen(file)
                    keyplus = key+'-'+str(dtype)
                    print('{0} read'.format(keyplus))
                    val = bgen.read(dtype=dtype).val
                    bgen.flush()
                    time_d = datetime.now()
                    list_row.append({'iid_count':iid_count,'sid_count':sid_count,'val_count':iid_count*sid_count,
                                     'bits':bits,'dtype':dtype,
                                     'read':time_d-time_c})
                    print(pd.DataFrame(list_row))
    return pd.DataFrame(list_row)
```
