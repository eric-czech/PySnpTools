```python
def default_sid_function(id,rsid):
    '''
    The default function for turning a Bgen (SNP) id and rsid into a :attr:`pysnptools.distreader.DistReader.sid`.
    If the Bgen rsid is '' or '0', the sid will be the (SNP) id.
    Otherwise, the sid will be 'ID,RSID'

    >>> default_sid_function('SNP1','rs102343')
    'SNP1,rs102343'
    >>> default_sid_function('SNP1','0')
    'SNP1'
    '''
    if rsid=='0' or rsid=='':
        return id
    else:
        return id+','+rsid
```

```python
import numpy as np
sid_count = 1000*1000
id_list = np.array(['sid_{0}'.format(i) for i in range(sid_count)])
rsid_list = np.array(['rsid{0}'.format(i) for i in range(sid_count)])
```

```python
%%time
col = np.array([default_sid_function(id,rsid) for id,rsid in zip(id_list,rsid_list)])
len(col)
```
