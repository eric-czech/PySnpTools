```python
import itertools

iid_count_list = [4**i for i in range(20)]
sid_count_list = iid_count_list
iid_sid_count_list = [(iid,sid) for iid,sid in itertools.product(iid_count_list,sid_count_list)]
iid_sid_count_list = sorted(iid_sid_count_list,key=lambda pair:pair[0]*pair[1])
iid_sid_count_list[-5:]
```

```python
from pysnptools.distreader import DistGen, Bgen
import  pandas as pd
from datetime import datetime

def timing(start,end,bits_list,compression_list):
    #df = pd.DataFrame(columns=['iid_count', 'sid_count', 'gen','cache','read'])
    
    list_row = []
    file = r'/mnt/m/del34/temp{0}.bgen'
    for bits in bits_list:
        for compression in compression_list:
            for iid_count,sid_count in iid_sid_count_list:
                if iid_count < 10 or sid_count < 10:
                    continue
                val_count = iid_count * sid_count
                if val_count < start:
                    continue
                if val_count >= end:
                    break #assumes iid_sid_count_list is sorted by val_count
                key = '{0}x{1}-{2}-{3}'.format(iid_count,sid_count,bits,compression)
                time_a = datetime.now()
                distgen = DistGen(seed=0,iid_count=iid_count,sid_count=sid_count)
                print('{0} gen and write to bgen'.format(key))
                bgen = Bgen.write(file.format(key),distgen,bits=bits,compression=compression)
                time_b = datetime.now()
                print('{0} cache'.format(key))
                bgen.iid
                bgen.flush()
                time_c = datetime.now()
                print('{0} read'.format(key))
                val = bgen.read().val
                bgen.flush()
                time_d = datetime.now()
                list_row.append({'iid_count':iid_count,'sid_count':sid_count,'val_count':iid_count*sid_count,
                                 'bits':bits,'compression':compression,
                                 'gen':time_b-time_a,'cache':time_c-time_b,'read':time_d-time_c})
                print(pd.DataFrame(list_row))
    return pd.DataFrame(list_row)
```

```python
df10K = timing(0,10*1000,[8,16],['zlib',None])
df10K
```

```python
df100K = timing(10*1000,100*1000,[8,16],['zlib',None])
df100K
```

```python
df100K.to_csv(r'df100k.tsv', sep="\t", index=False)
```
